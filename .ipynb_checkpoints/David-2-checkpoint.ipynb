```python
evo=pd.read_csv(data_dir + "evo-est.csv", delimiter=";",encoding="latin1")
evo=make_cols_numeric(evo,['Població estrangera. % sobre (1)'])
```

```python
from matplotlib.ticker import FuncFormatter
x=evo['Categoria']
y1=evo['Població estrangera. % sobre (1)']
y2=evo['Població estrangera. Total']
fig, ax1 = plt.subplots()

# Plot data for the left y-axis
line1=ax1.scatter(x, y1, label='Foreign Population (%)',c='darkblue', marker='s', edgecolor='black', linewidth=0.5)
ax1.plot(x, y1, linestyle='-', color='darkblue', linewidth=1)
ax1.set_xlabel('Year')
ax1.set_ylabel('Foreign Population, over total population (%)')
ax1.tick_params(axis='y')

# Create the secondary axis (sharing the same x-axis)
ax2 = ax1.twinx()
line2=ax2.scatter(x, y2, color='none', marker='o', label='Foreign Population (Absolute)', edgecolor='red', linewidth=1)
ax2.plot(x, y2, linestyle='-', color='red', linewidth=1, label='Foreign Population (Absolute)')
ax2.set_ylabel('Total Foreign Population (x$10^6$)')
ax2.tick_params(axis='y')

def short_format(x, _):
#    if x >= 1e6:
        return f"{x / 1e6:.1f}"  # Millions with 1 decimal

ax2.yaxis.set_major_formatter(FuncFormatter(short_format))

# Add legends for both axes
handles = [line1, line2]
labels = [h.get_label() for h in handles]

# Create a single legend
ax1.legend(handles, labels, loc='lower right', ncol=1)
plt.tight_layout()
plt.savefig(data_dir+'evo-art.png',bbox_inches='tight',dpi=600, transparent=False)
plt.show()
```

```python
x1=merged_p['% sap parlar']
y1=merged2_p['Població estrangera. % sobre (1)']
x2=merged3_p['% sap parlar']
y2=merged4_p['Població estrangera. % sobre (1)']
y=x1-x2
x=y1-y2
from scipy.stats import linregress

marker1='o'
marker2='v'
marker3='s'

slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_pred = [slope * xi + intercept for xi in x]
residuals = np.array(y) - np.array(y_pred)
std_dev = np.std(residuals)
x_line = np.linspace(5, 22.5, 500)
y_line = slope * x_line + intercept
plt.scatter(x, y, label='Difference between 2001 and 2011 data ',
            #c='darkblue', 
            c=x, cmap='RdYlBu_r',vmin=6,vmax=22,
            marker=marker1,zorder=2, edgecolor='black', linewidth=0.5)
plt.plot(x_line, y_line, 
         label=f'Linear Regression (R²={r_value**2:.2f}, $\sigma$ =±{std_dev:.2f})',
         color='red',zorder=1)
plt.fill_between(x_line, y_line - std_dev, y_line + std_dev, color='red', alpha=0.2,
                 #label=f'$\sigma$ (±{std_dev2:.2f})',
                zorder=0)

# Add labels and title
plt.ylabel('Variation of Catalan Speakers (%)')
plt.xlabel('Variation of Foreign Population (%)')
# Add a legend
plt.legend(
    ncol=1, bbox_to_anchor=(0, 0, 1, 1))
plt.xlim(5, 22.5)

plt.tight_layout()
plt.savefig(data_dir+'correlation3-art.png',bbox_inches='tight',dpi=600, transparent=False)
plt.show()
```

```python
print(f"P-value: {p_value}")
print(f"Standard dev: {std_dev}")
print(f"R^2: {r_value**2}")
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpecFromSubplotSpec

# Create a figure and specify the grid layout
fig = plt.figure(figsize=(15, 18))  # Adjust size as needed
#gs = GridSpec(2, 2, figure=fig, height_ratios=[1, 1, 2])  # 3 rows, single column
gs = GridSpec(2, 2, figure=fig, height_ratios=[1, 2], width_ratios=[1, 1])  # 2 rows, 2 columns
# ------------------ FIRST PLOT ------------------
xa=evo['Categoria']
ya=evo['Població estrangera. % sobre (1)']
ax1 = fig.add_subplot(gs[0, 0])  # First subplot
ax1.scatter(xa, ya, label='Foreign Population (%)', c='darkblue', marker='s', edgecolor='black', linewidth=0.5)
ax1.plot(xa, ya, linestyle='-', color='darkblue', linewidth=1)
ax1.set_xlabel('Year')
ax1.set_ylabel('Foreign Population, over total population (%)')
ax1.tick_params(axis='y')
ax1.legend(loc='lower right')
ax1.set_title('a.')

# ------------------ SECOND PLOT ------------------
x1=merged_p['% sap parlar']
y1=merged2_p['Població estrangera. % sobre (1)']
x2=merged3_p['% sap parlar']
y2=merged4_p['Població estrangera. % sobre (1)']
y=x1-x2
x=y1-y2
from scipy.stats import linregress
marker1='o'
slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_pred = [slope * xi + intercept for xi in x]
residuals = np.array(y) - np.array(y_pred)
std_dev = np.std(residuals)
x_line = np.linspace(5, 22.5, 500)
y_line = slope * x_line + intercept
ax2 = fig.add_subplot(gs[0,1])  # Second subplot
ax2.scatter(x, y, label='Difference between 2001 and 2011 data', c=x, cmap='RdYlBu_r', vmin=6, vmax=22,
            marker=marker1, zorder=2, edgecolor='black', linewidth=0.5)
ax2.plot(x_line, y_line, label=f'Linear Regression (R²={r_value**2:.2f}, σ =±{std_dev:.2f})', color='red', zorder=1)
ax2.fill_between(x_line, y_line - std_dev, y_line + std_dev, color='red', alpha=0.2, zorder=0)
ax2.set_xlabel('Variation of Foreign Population (%)')
ax2.set_ylabel('Variation of Catalan Speakers (%)')
ax2.legend(loc='upper left')
ax2.set_xlim(5, 22.5)
ax2.set_title('b.')

# ------------------ THIRD PLOT ------------------
# Create a nested GridSpec for the map grid
ax3 = fig.add_subplot(gs[1, :])
map_grid = GridSpecFromSubplotSpec(2, 2, subplot_spec=gs[1,:], wspace=0.4, hspace=0.)

# Add each map plot to the nested grid
ax1986_00 = fig.add_subplot(map_grid[0, 0])
merged3_p.plot(column='% sap parlar', cmap='RdYlBu_r', legend=True, ax=ax1986_00, 
               legend_kwds={"label": "Percentage (%)","orientation": "vertical", 'shrink': 0.8}, vmin=63, vmax=93)
ax1986_00.set_title('Catalan Speakers (2001)')
ax1986_00.axis('off')

ax1986_10 = fig.add_subplot(map_grid[1, 0])
merged4_p.plot(column='Població estrangera. % sobre (1)', cmap='RdYlBu_r', legend=True, ax=ax1986_10, 
               legend_kwds={"label": "Percentage (%)","orientation": "vertical", 'shrink': 0.8}, vmin=1, vmax=28)
ax1986_10.set_title('Foreign Population (2001)')
ax1986_10.axis('off')

ax1986_01 = fig.add_subplot(map_grid[0, 1])
merged_p.plot(column='% sap parlar', cmap='RdYlBu_r', legend=True, ax=ax1986_01, 
              legend_kwds={"label": "Percentage (%)","orientation": "vertical", 'shrink': 0.8}, vmin=63, vmax=93)
ax1986_01.set_title('Catalan Speakers (2011)')
ax1986_01.axis('off')

ax1986_11 = fig.add_subplot(map_grid[1, 1])
merged2_p.plot(column='Població estrangera. % sobre (1)', cmap='RdYlBu_r', legend=True, ax=ax1986_11, 
               legend_kwds={"label": "Percentage (%)","orientation": "vertical", 'shrink': 0.8}, vmin=1, vmax=28)
ax1986_11.set_title('Foreign Population (2011)')
ax1986_11.axis('off')
ax3.set_title('c.')
ax3.tick_params(axis='both', which='both', length=0, labelleft=False, labelbottom=False)
# Add global annotations
fig.text(0.055, 0.015, '* Over respective Comarca population.', fontsize=9)

# ------------------ SAVE AND SHOW ------------------
plt.tight_layout()
plt.savefig(data_dir + 'combined_figure.png', bbox_inches='tight', dpi=600, transparent=False)
plt.show()
```
