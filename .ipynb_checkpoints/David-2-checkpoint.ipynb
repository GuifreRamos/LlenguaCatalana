```python
import numpy as np
import pandas as pd
data_dir = "../LlenguaCatalana/dades_david/"


dat1 = pd.read_csv(data_dir + "cat2011.csv", delimiter=";")
dat2 = pd.read_csv(data_dir + "cat2001.csv", delimiter=";")
dat3 = pd.read_csv(data_dir + "est2011-2.csv", delimiter=";",encoding="latin1")
dat4 = pd.read_csv(data_dir + "est2001.csv", delimiter=";",encoding="latin1")

# show loaded table:
#dat4
```

```python
import geopandas as gpd
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data_dir = "../LlenguaCatalana/dades_david/"
comarques = gpd.read_file(data_dir + "comarq.geojson")
```

```python
dat1.rename(columns={'Unnamed: 0': 'nom_comar'}, inplace=True)
```

```python
merged = comarques.merge(dat1, on="nom_comar")
```

```python
dat3.rename(columns={'Categoria': 'nom_comar'}, inplace=True)
```

```python
merged2 = comarques.merge(dat3, on="nom_comar")
```

```python
dat2.rename(columns={'Unnamed: 0': 'nom_comar'}, inplace=True)
```

```python
merged3 = comarques.merge(dat2, on="nom_comar")
```

```python
dat4.rename(columns={'Categoria': 'nom_comar'}, inplace=True)
```

```python
merged4 = comarques.merge(dat4, on="nom_comar")
```

```python
from Scripts.Lucas.utils import make_cols_numeric
merged_p=make_cols_numeric(merged,['% sap parlar'])
merged2_p=make_cols_numeric(merged2,['Població estrangera. % sobre (1)'])
```

```python
from Scripts.Lucas.utils import make_cols_numeric
merged3_p=make_cols_numeric(merged3,['% sap parlar'])
merged4_p=make_cols_numeric(merged4,['Població estrangera. % sobre (1)'])
```

```python
evo=pd.read_csv(data_dir + "evo-est.csv", delimiter=";",encoding="latin1")
evo=make_cols_numeric(evo,['Població estrangera. % sobre (1)'])
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec, GridSpecFromSubplotSpec

# Create a figure and specify the grid layout
fig = plt.figure(figsize=(14, 7))  # Adjust size as needed
gs = GridSpec(2, 2, figure=fig, height_ratios=[1, 1.75], width_ratios=[1, 2])  # 2 rows, 2 columns

# ------------------ FIRST PLOT ------------------
xa = evo['Categoria']
ya = evo['Població estrangera. % sobre (1)']
ax1 = fig.add_subplot(gs[0, 0])  # First subplot
ax1.scatter(xa, ya, label='Foreign Population (%)', c='darkblue', marker='s', edgecolor='black', linewidth=0.5)
ax1.plot(xa, ya, linestyle='-', color='darkblue', linewidth=1)
ax1.set_xlabel('Year',fontsize=10)
ax1.set_ylabel('Total Foreign Population (%)',fontsize=10)
ax1.tick_params(axis='y')
ax1.legend(loc='lower right',fontsize=10)
ax1.set_title('a.', fontweight='bold')

# ------------------ SECOND PLOT ------------------
x1 = merged_p['% sap parlar']
y1 = merged2_p['Població estrangera. % sobre (1)']
x2 = merged3_p['% sap parlar']
y2 = merged4_p['Població estrangera. % sobre (1)']
y = x1 - x2
x = y1 - y2
from scipy.stats import linregress
marker1 = 'o'
slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_pred = [slope * xi + intercept for xi in x]
residuals = np.array(y) - np.array(y_pred)
std_dev = np.std(residuals)
x_line = np.linspace(5, 22.5, 500)
y_line = slope * x_line + intercept
ax2 = fig.add_subplot(gs[1, 0])  # Second subplot
ax2.scatter(x, y, label=r'$Data[2001]-Data[2011]$', c=x, cmap='RdYlBu_r', vmin=6, vmax=22,
            marker=marker1, zorder=2, edgecolor='black', linewidth=0.5)
ax2.plot(x_line, y_line, label=f'Regression (R²={r_value**2:.2f}, σ =±{std_dev:.2f})', color='red', zorder=1)
ax2.fill_between(x_line, y_line - std_dev, y_line + std_dev, color='red', alpha=0.2, zorder=0)
ax2.set_xlabel('Variation of Foreign Population (%)',fontsize=10)
ax2.set_ylabel('Variation of Catalan Speakers (%)',fontsize=10)
ax2.legend(loc='upper right', bbox_to_anchor=(1, 1),fontsize=10)
ax2.set_xlim(5, 22.5)
ax2.set_title('c.', fontweight='bold')

# ------------------ THIRD PLOT ------------------
# Create a nested GridSpec for the map grid in the bottom right
ax3 = fig.add_subplot(gs[:, 1])  # Third subplot that spans the second row, second column
map_grid = GridSpecFromSubplotSpec(2, 2, subplot_spec=gs[:, 1], wspace=0., hspace=0.)

# Add each map plot to the nested grid
ax1986_00 = fig.add_subplot(map_grid[0, 0])
merged3_p.plot(column='% sap parlar', cmap='RdYlBu_r', legend=True, ax=ax1986_00,
               legend_kwds={"label": "Percentage (%)", "orientation": "vertical", 'shrink': 0.8}, vmin=63, vmax=93)
ax1986_00.set_title('Catalan Speakers (2001)', fontsize=10,x=0.7, y=0.075)
ax1986_00.axis('off')

ax1986_10 = fig.add_subplot(map_grid[1, 0])
merged4_p.plot(column='Població estrangera. % sobre (1)', cmap='RdYlBu_r', legend=True, ax=ax1986_10,
               legend_kwds={"label": "Percentage (%)", "orientation": "vertical", 'shrink': 0.8}, vmin=1, vmax=28)
ax1986_10.set_title('Foreign Population (2001)', fontsize=10,x=0.7, y=0.075)
ax1986_10.axis('off')

ax1986_01 = fig.add_subplot(map_grid[0, 1])
merged_p.plot(column='% sap parlar', cmap='RdYlBu_r', legend=True, ax=ax1986_01,
              legend_kwds={"label": "Percentage (%)", "orientation": "vertical", 'shrink': 0.8}, vmin=63, vmax=93)
ax1986_01.set_title('Catalan Speakers (2011)', fontsize=10,x=0.7, y=0.075)
ax1986_01.axis('off')

ax1986_11 = fig.add_subplot(map_grid[1, 1])
merged2_p.plot(column='Població estrangera. % sobre (1)', cmap='RdYlBu_r', legend=True, ax=ax1986_11,
               legend_kwds={"label": "Percentage (%)", "orientation": "vertical", 'shrink': 0.8}, vmin=1, vmax=28)
ax1986_11.set_title('Foreign Population (2011)', fontsize=10,x=0.7, y=0.075)
ax1986_11.axis('off')

# Add title for the third plot
ax3.set_title('b.', fontweight='bold')
ax3.tick_params(axis='both', which='both', length=0, labelleft=False, labelbottom=False)

#fig.text(0.7, 0.05, 'Percentages are over respective Comarca population.', fontsize=10)
# ------------------ SAVE AND SHOW ------------------
plt.tight_layout()
plt.savefig(data_dir +'combined_figure_v2.png', bbox_inches='tight', dpi=600, transparent=False)
plt.show()
```
